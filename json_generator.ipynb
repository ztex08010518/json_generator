```python
class globalProfile():
    def __init__(self): 
        self.description = ""
        self.profileID = random.randint(0,9)
        self.ipList = IpsIDs(5)
        self.catList = CatList(4)
        
    def globalProfile_vpWhiteList(self):
        self.content = {
            "description" : "%s" % self.description,
            "profileID" : "%d" % self.profileID,
            "IpsIDs" : []
        }
        self.content["IpsIDs"].extend(self.ipList.IDList)    
    def globalProfile_appProtal(self):
        self.content = {}
    def globalProfile_webReputationService(self):
        self.content = {
            "profileID" : "%d" %self.profileID,
            "catList" : []
        }
        self.content["catList"].extend(self.catList.catlist)
    def globalProfile_urlWhiteBlackList(self):
        tmp_bl = blackList()
        tmp_bl.blackList_url_glo()
        tmp_wl = whiteList()
        tmp_wl.whiteList_url_glo()
        self.content = {
            "profileID" : "%d" %random.randint(0,9),
            "listPriority" : "blackList",
            "blackList" : {},
            "whiteList" : {}
        }
        self.content["blackList"] = (tmp_bl.content)
        self.content["whiteList"] = (tmp_wl.content)
    def globalProfile_appWhiteBlackList(self):
        self.content = {
            "description" : "",
            "profileID" : "%d" %random.randint(0,9),
            "listPriority" : "blackList",
            "blackList" : [],
            "whiteList" : []
        }
        tmp_app_1 = appList(2)
        tmp_app_2 = appList(1)
        self.content["blackList"].extend(tmp_app_1.applist)
        self.content["whiteList"].extend(tmp_app_2.applist)
    def globalProfile_appAuth(self):
        tmp_cert = certAppList(2)
        tmp_rel = relAppList(3)
        time = 60*random.randint(0,9)
        self.content = {
            "allowPeriod" : "%d" %time,
            "certAppList" : [],
            "relAppList" : []
        }
        self.content["certAppList"].extend(tmp_cert.certapplist)
        self.content["relAppList"].extend(tmp_rel.relapplist)
```

```python
def rand_mac():
    return "%02x:%02x:%02x:%02x:%02x:%02x" %(
    random.randint(0,255),
    random.randint(0,255),
    random.randint(0,255),
    random.randint(0,255),
    random.randint(0,255),
    random.randint(0,255)
    )
```

```python
def rand_ID_8():
    return "%08d" %(random.randint(0,99999999))
```

```python
def ip():
    return "%03d.%03d.%03d.%03d" %(
    random.randint(0,999),
    random.randint(0,999),
    random.randint(0,999),
    random.randint(0,999)
    )
```
